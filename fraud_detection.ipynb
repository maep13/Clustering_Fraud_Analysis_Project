## Preprocesamiento

```python
# Código para transformar las columnas que tienen rangos de valor como entradas

# Mapeo para AgeOfVehicle

age_vehicle_map = {
    "3 years": 3,
    "5 years": 5,
    "6 years": 6,
    "7 years": 7,
    "more than 7": 8  # Asumimos que 'más de 7 años' se representa como 8
}

# Mapeo para AgeOfPolicyHolder

age_policy_map = {
    "16 to 17": 16.5,
    "18 to 20": 19,
    "21 to 25": 23,
    "26 to 30": 28,
    "31 to 35": 33,
    "36 to 40": 38,
    "41 to 50": 45.5,
    "51 to 65": 58,
    "over 65": 70  # Asumimos que 'más de 65' es representado como 70
}

# Transformar AgeOfVehicle

df['AgeOfVehicle'] = df['AgeOfVehicle'].map(age_vehicle_map)

# Transformar AgeOfPolicyHolder

df['AgeOfPolicyHolder'] = df['AgeOfPolicyHolder'].map(age_policy_map)

# Verificar las transformaciones

print("Transformación de AgeOfVehicle:")
print(df['AgeOfVehicle'].head())
print("Transformación de AgeOfPolicyHolder:")
print(df['AgeOfPolicyHolder'].head())
```

```python
# Mapeo para Days_Policy_Accident

days_policy_accident_map = {
    "none": 0,
    "1 to 7": 4,
    "8 to 15": 11.5,
    "15 to 30": 22.5,
    "more than 30": 35
}

# Mapeo para PastNumberOfClaims

past_claims_map = {
    "none": 0,
    "1": 1,
    "2 to 4": 3,
    "more than 4": 5
}

# Mapeo para NumberOfSuppliments

supplements_map = {
    "none": 0,
    "1 to 2": 1.5,
    "more than 2": 3
}

# Transformar las columnas

df['Days_Policy_Accident'] = df['Days_Policy_Accident'].map(days_policy_accident_map)
df['PastNumberOfClaims'] = df['PastNumberOfClaims'].map(past_claims_map)
df['NumberOfSuppliments'] = df['NumberOfSuppliments'].map(supplements_map)

# Verificar las transformaciones

print("Transformaciones realizadas:")
print(df[['Days_Policy_Accident', 'PastNumberOfClaims', 'NumberOfSuppliments']].head())
```

```python
df.info()
```

```python
# Mapeo para numberofcars

# Mapeo para numberofcars
number_of_cars_map = {
    '1 vehicle': 1,
    '2 vehicles': 2,
    '3 to 4': 3.5,    # Promedio entre 3 y 4
    '5 to 8': 6.5,    # Promedio entre 5 y 8
    'more than 8': 9  # Representamos "more than 8" como 9
}

# Aplicar el mapeo

df['NumberOfCars'] = df['NumberOfCars'].map(number_of_cars_map)

# Mapeo para addresschange_claim

address_change_map = {
    'no change': 0,
    'under 6 months': 0.5,  # Representamos "menos de 6 meses" con un valor bajo
    '1 year': 1,
    '2 to 3 years': 2.5,  # Promedio entre 2 y 3 años
    '4 to 8 years': 6,  # Promedio entre 4 y 8 años
}
df['AddressChange_Claim'] = df['AddressChange_Claim'].map(address_change_map)

# Mapeo para days_policy_claim

days_policy_claim_map = {
    'none': 0,
    '1 to 7': 4,
    '8 to 15': 11.5,  # Promedio entre 8 y 15
    '15 to 30': 22.5,  # Promedio entre 15 y 30
    'more than 30': 35  # Representamos "más de 30 días" como 35
}
df['Days_Policy_Claim'] = df['Days_Policy_Claim'].map(days_policy_claim_map)

# Revisar y eliminar valores problemáticos en monthclaimed ('0')

df = df[df['MonthClaimed'] != '0']

# Aplicar one-hot encoding para monthclaimed

df = pd.get_dummies(df, columns=['MonthClaimed'], drop_first=True)

# Verificar las transformaciones realizadas

print("Transformación de numberofcars:")
print(df['NumberOfCars'].head())

print("Transformación de addresschange_claim:")
print(df['AddressChange_Claim'].head())

print("Transformación de days_policy_claim:")
print(df['Days_Policy_Claim'].head())

print("Columnas después del one-hot encoding para MonthClaimed:")
print(df.columns)
```

```python
# Imprimir valores nulos de cada columna

print("Valores nulos por columna:")
print(df.isnull().sum())
```

```python
# Rellenar valores faltantes "NaN" con la mediana

median_value = df['NumberOfSuppliments'].median()
df['NumberOfSuppliments'] = df['NumberOfSuppliments'].fillna(median_value)
median_value = df['AgeOfVehicle'].median()
df['AgeOfVehicle'] = df['AgeOfVehicle'].fillna(median_value)
```

```python
# Definir las columnas categóricas que queremos transformar a nuevas columnas binarias

categorical_columns = [
    'Month', 'DayOfWeek', 'AccidentArea', 'DayOfWeekClaimed', 'Sex',
    'MaritalStatus', 'Fault', 'PolicyType', 'VehicleCategory', 'VehiclePrice',
    'PoliceReportFiled', 'WitnessPresent', 'AgentType', 'BasePolicy', 'Make'
]

# Aplicar one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

```python
# Estandarizar nombres de columnas

df_encoded.columns = (
    df_encoded.columns
    .str.strip()                # Elimina espacios al inicio y final
    .str.lower()                # Convierte a minúsculas
    .str.replace(' ', '_')      # Reemplaza espacios con guiones bajos
    .str.replace('-', '_')      # Reemplaza guiones con guiones bajos
    .str.replace('.', '_')      # Reemplaza puntos con guiones bajos
)
```

```python
# Verificar las nuevas columnas generadas

print("Nuevas columnas después del one-hot encoding:")
print(df_encoded.columns)
```

```python
# criterio basado en datos para descartar características 

# Columnas con valores casi constantes (baja varianza) aportan poca o ninguna información útil al modelo. 
# Este método es ideal para eliminar esas características redundantes.

# Filtrar características con una varianza baja

selector = VarianceThreshold(threshold=0.01)  # Umbral ajustable (0.01 para baja variabilidad)
reduced_features = selector.fit_transform(df_encoded)

# Mostrar las columnas retenidas
selected_columns = df_encoded.columns[selector.get_support()]
print("Columnas retenidas por varianza:")
print(selected_columns)


# Si dos características están altamente correlacionadas, el modelo puede funcionar igual de bien con una sola de ellas. Por lo tanto, 
# podemos descartar una para evitar redundancia.

# Calcular matriz de correlación
correlation_matrix = df_encoded.corr()

# Identificar características con correlación mayor a un umbral
threshold = 0.8  # Puedes ajustar este valor
high_corr = np.where(np.abs(correlation_matrix) > threshold)
high_corr_pairs = [(correlation_matrix.index[x], correlation_matrix.columns[y]) 
                   for x, y in zip(*high_corr) if x != y]

print("Pares de columnas altamente correlacionadas:")
print(high_corr_pairs)

# Elimina una columna por cada par correlacionado alto
```

```python
# Usa pruebas como ANOVA o Chi-cuadrado para medir la relación de cada característica con la variable objetivo.
# Esto es especialmente útil para datos categóricos.

# Selección de las mejores características según ANOVA

# Definir las características (todas menos 'fraudfound_p') y la variable objetivo

features = df_encoded.drop(columns=['fraudfound_p'])  # Excluye la variable objetivo
target = df_encoded['fraudfound_p']  # Variable objetivo

selector = SelectKBest(score_func=f_classif, k=10)  # Selecciona las 10 mejores características
best_features = selector.fit_transform(features, target)

# Mostrar columnas seleccionadas
selected_columns = features.columns[selector.get_support()]
print("Columnas seleccionadas según ANOVA:")
print(selected_columns)
```

```python
# Dividir el dataset en características y variable objetivo

features = df_encoded.drop(columns=['fraudfound_p'])  # Excluye la variable objetivo
target = df_encoded['fraudfound_p']  # Variable objetivo

# Dividir en conjuntos de entrenamiento y prueba

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42, stratify=target)

print("Tamaño del conjunto de entrenamiento:", X_train.shape)
print("Tamaño del conjunto de prueba:", X_test.shape)
```

```python
# Crear el objeto SMOTE

smote = SMOTE(random_state=42)

# Aplicar SMOTE al conjunto de entrenamiento

X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print("Tamaño del conjunto de entrenamiento después de SMOTE:", X_train_smote.shape)
print("Distribución de clases después de SMOTE:")
print(y_train_smote.value_counts())
```

```python
# Crear el objeto de normalización

scaler = StandardScaler()

# Ajustar el scaler en los datos de entrenamiento y transformar

X_train_scaled = scaler.fit_transform(X_train_smote)  # Ajuste basado en los datos de entrenamiento balanceados
X_test_scaled = scaler.transform(X_test)  # Transformar los datos de prueba sin ajuste adicional

print("Forma del entrenamiento escalado:", X_train_scaled.shape)
print("Forma del conjunto de prueba escalado:", X_test_scaled.shape)
```

```python
# 1. Verificar la distribución de clases después de SMOTE

print("Distribución de clases después de SMOTE:")
print(pd.Series(y_train_smote).value_counts(normalize=True))

# 2. Inspeccionar los valores escalados

print("\nMedia de las características escaladas en el conjunto de entrenamiento:")
print(np.mean(X_train_scaled, axis=0))

print("\nDesviación estándar de las características escaladas en el conjunto de entrenamiento:")
print(np.std(X_train_scaled, axis=0))

# 3. Verificar consistencia entre entrenamiento y prueba

assert X_train_scaled.shape[1] == X_test_scaled.shape[1], "Las columnas en entrenamiento y prueba no coinciden."
print("\nConjuntos consistentes: entrenamiento y prueba tienen las mismas características.")

# 4. Evaluación de valores extremos

print("\nValores mínimos en el conjunto de entrenamiento escalado:")
print(np.min(X_train_scaled, axis=0))

print("\nValores máximos en el conjunto de entrenamiento escalado:")
print(np.max(X_train_scaled, axis=0))

# 5. Comprobar valores nulos

print("\nValores nulos en los datos escalados:")
print(pd.DataFrame(X_train_scaled).isnull().sum().sum())
```

```python
# Crear el modelo de Random Forest

model = RandomForestClassifier(random_state=42)

# Métricas que vamos a evaluar

scoring = ['accuracy', 'precision', 'recall', 'f1']

# Realizar K-fold cross-validation

k = 5  # Número de folds
cv_results = cross_validate(model, X_train_scaled, y_train_smote, cv=k, scoring=scoring)

# Resultados: Accuracy, Precision, Recall y F1-score en cada fold

print("Accuracy en cada fold:", cv_results['test_accuracy'])
print(f"Promedio Accuracy: {np.mean(cv_results['test_accuracy']):.4f}")
print("Precision en cada fold:", cv_results['test_precision'])
print(f"Promedio Precision: {np.mean(cv_results['test_precision']):.4f}")
print("Recall en cada fold:", cv_results['test_recall'])
print(f"Promedio Recall: {np.mean(cv_results['test_recall']):.4f}")
print("F1-score en cada fold:", cv_results['test_f1'])
print(f"Promedio F1-score: {np.mean(cv_results['test_f1']):.4f}")
```

## Segmentación de Características más relevantes

```python
# Escalar todos los datos disponibles en el DataFrame

segmentation_features = df_encoded.iloc[:, :]  # Tomamos todas las columnas
scaler = StandardScaler()
scaled_features = scaler.fit_transform(segmentation_features)

# Aplicar clustering jerárquico

model = AgglomerativeClustering(n_clusters=5, linkage='ward')  # Cambia n_clusters según sea necesario
clusters = model.fit_predict(scaled_features)

# Añadir clusters al DataFrame

df_encoded['Cluster'] = clusters
```

```python
# Reducir a 3 componentes principales
#pca = PCA(n_components=2)
#pca_features = pca.fit_transform(scaled_features)

# Graficar los clusters en 3D
#fig = plt.figure(figsize=(10, 8))
#ax = fig.add_subplot(111, projection='3d')
#ax.scatter(pca_features[:, 0], pca_features[:, 1], pca_features[:, 2], c=clusters, cmap='viridis', s=50)
#ax.set_xlabel('Componente PCA 1')
#ax.set_ylabel('Componente PCA 2')
#ax.set_zlabel('Componente PCA 3')
#ax.set_title('Clusters visualizados en 3D')
#plt.show()

# Reducir a dos componentes principales
pca = PCA(n_components=30)
pca_features = pca.fit_transform(scaled_features)  # scaled_features: datos escalados

# Visualizar los clusters en 2D
plt.figure(figsize=(10, 7))
plt.scatter(pca_features[:, 0], pca_features[:, 1], c=df_encoded['Cluster'], cmap='viridis', s=50)
plt.title('Gráfica 2D basada en PCA')
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Aplicar Agglomerative Clustering

model = AgglomerativeClustering(n_clusters=2, linkage='average')  # Cambia n_clusters según lo necesario
clusters = model.fit_predict(pca_features)  # pca_features contiene las 6 componentes principales

# Mostrar resultados

print(f"Clusters encontrados: {len(set(clusters))}")

# Visualizar dendrograma

linked = linkage(pca_features, method='ward')  # 'ward' minimiza la varianza dentro de cada cluster
plt.figure(figsize=(10, 7))
dendrogram(linked, truncate_mode='level', p=5)  # Reduce detalles con truncate_mode
plt.title('Dendrograma del clustering jerárquico')
plt.xlabel('Puntos de datos')
plt.ylabel('Distancia')
plt.show()
```

```python
# Calculo de Silhouette Score

score = silhouette_score(pca_features, clusters)
print(f"Silhouette score: {score:.4f}")
```

```python
# Calculo de Varianza

explained_variance = pca.explained_variance_ratio_
print("Varianza explicada por las componentes principales:", explained_variance.sum())
```

```python
# Añadir los clusters al DataFrame

df_encoded['Cluster'] = clusters

# Calcular el promedio de características por cluster

cluster_means = df_encoded.groupby('Cluster').mean()
print("Promedio de características por cluster:")

pd.set_option('display.max_columns', None)  # Muestra todas las columnas
pd.set_option('display.max_rows', None)     # Muestra todas las filas (opcional)

print(cluster_means)
```
